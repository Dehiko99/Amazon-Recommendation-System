## Simple Collaborative Filtering

```python
books_features_df = popular_books.pivot_table(index="Book-Title",columns="User-ID", values="Book-Rating").fillna(0)
books_features_df.head(10)
```

```python
from scipy.sparse import csr_matrix
books_features_df_matrix = csr_matrix(books_features_df.values)

from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric="cosine", algorithm="brute")
model_knn.fit(books_features_df_matrix)
```

```python
query_index = np.random.choice(books_features_df.shape[0])
print(query_index)
```

```python
distances, indices = model_knn.kneighbors(books_features_df.iloc[query_index,:].values.reshape(1,-1), n_neighbors = 11)
```

```python
for i in range(0, len(distances.flatten())):
  if i == 0:
    print("Recommendations for {0}: \n".format(books_features_df.index[query_index]))
  else:
    print("{0}:{1}, with distance of {2}".format(i,books_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```

I want to create an advanced Neural Collaborative Filtering (NCF), incorporating additional features (like author, location, age). The traditional NCF uses just **user embeddings** and **item embeddings** to learn the relationship between users and items. To incorporate additional features, I can trate them as additional inputs to the neural network. The embeddings of these features can then be concatenated with the user and item embeddings before passing them through the neural network layers.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F  
```

```python
X_train_explicit.columns
```

**Creation of the Neural Network**

```python
class NCF_Explicit_Feedback(nn.Module):
    def __init__(self, n_users, n_items,  embedding_dim):
        super(NCF_Explicit_Feedback, self).__init__() 
        self.user_embedding = nn.Embedding(n_users, embedding_dim)
        self.item_embedding = nn.Embedding(n_items, embedding_dim)

        self.fc1 = nn.Linear(embedding_dim * 2, 128)  # First hidden layer
        self.fc2 = nn.Linear(128, 64)                      # Second hidden layer
        self.fc3 = nn.Linear(64,1)                         # Additional hidden layer


        # Optionally add a dropout layer to prevent overfitting
        self.dropout = nn.Dropout(p=0.4)  # Dropout with 40% probability
        self.init_weights()

    def init_weights(self):
        # Use Xavier initialization
        for layer in [self.fc1, self.fc2, self.fc3]:
            nn.init.xavier_uniform_(layer.weight)
            layer.bias.data.fill_(0.01)

    def forward(self, user_input, item_input):
        #Flatten categorical variables
        user_vector = self.user_embedding(user_input).view(user_input.size(0),-1)
        item_vector = self.item_embedding(item_input).view(item_input.size(0),-1)

        #age and year are already continuous, no need to flatten
        concatenation = torch.cat([user_vector,item_vector], dim= -1)

        output = self.fc1(concatenation)
        output = F.relu(output)
        output = self.dropout(output)

        output = self.fc2(output)
        output = F.relu(output)  
        output = self.dropout(output)

        output = self.fc3(output)

        return output
```

```python
n_users = new_data_filtrato["User-ID"].nunique()
n_items = new_data_filtrato["ISBN"].nunique()
```

```python
train_users_tensors = torch.tensor(X_train_explicit["User-ID"].values)
train_items_tensors = torch.tensor(X_train_explicit["ISBN"].values)
train_explicit_ratings_tensors = torch.tensor(y_train_explicit_normalized, dtype=torch.float32)

test_users_tensors = torch.tensor(X_test_explicit["User-ID"].values)
test_items_tensors = torch.tensor(X_test_explicit["ISBN"].values)
test_explicit_ratings_tensors = torch.tensor(y_test_explicit_normalized, dtype=torch.float32)
```

```python
batch_size = 128
from torch.utils.data import DataLoader, TensorDataset
train_dataset = TensorDataset(train_users_tensors, train_items_tensors, train_explicit_ratings_tensors)
train_loader = DataLoader(train_dataset,batch_size=batch_size, shuffle=True, num_workers=4)
```

```python
test_dataset = TensorDataset(test_users_tensors, test_items_tensors, test_explicit_ratings_tensors)
test_loader = DataLoader(test_dataset, batch_size=batch_size,shuffle=False)
```

**Training and testing the model**

```python
embedding_dim = 32
model = NCF_Explicit_Feedback(n_users=n_users, n_items=n_items, embedding_dim=embedding_dim)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.00001,weight_decay=1e-6)  
```

```python
num_epochs = 20
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for user_batch, item_batch, explicit_ratings_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(user_batch,item_batch)
        loss = criterion(outputs,explicit_ratings_batch)
        loss.backward()
        optimizer.step()
        total_loss += total_loss
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader)}")
```

```python
model.eval()  # Set the model to evaluation mode
    
total_loss = 0
total_samples = 0

with torch.no_grad():  # No need to calculate gradients during evaluation
    for batch in test_loader:
        user_input, item_input, ratings = batch
            
            # Forward pass: Get predictions from the model
        predictions = model(user_input, item_input)
            
            # Calculate MSE Loss
        loss = criterion(predictions, ratings)
            
            # Accumulate the total loss
        total_loss += loss.item() * len(ratings)  # Multiply by batch size
        total_samples += len(ratings)
avg_loss = total_loss / total_samples
print(f"Test MSE: {avg_loss:.4f}")
```
